```python
%matplotlib inline
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = fits.open('try1.fits')
file.info()
data = file[0].header
```

```python
data
```

```python
file[1].header
```

```python
#read flux, wavelength and redshift
data2 = file[1].data
flux = data2['flux']
wave = 10**data2['loglam']

plt.plot(wave,flux)
plt.xlabel('Wavelength')
plt.ylabel('Flux')
```

```python
#redshift
data2 = file[1].data
flux = data2['flux']
wave = 10**data2['loglam']

plt.plot(wave,flux)

z = file[2].data['z']
z = z[0]
plt.title('Spectra Star')
plt.xlabel(r'wavelength $ \AA$')
plt.ylabel(r'flux $ 10^{-17} erg/cm^2/s/ \AA$')
```

##SPECTRA IN THE REST FRMAE FOR Z = 0.0288

```python
plt.figure()
plt.plot(wave/(1+z),flux,color='red')
plt.title('Spectra Star')
plt.xlabel(r'wavelength $ \AA$')
plt.ylabel(r'flux $ 10^{-17} erg/cm^2/s/ \AA$')
plt.axvline(x=6564.16)
```
